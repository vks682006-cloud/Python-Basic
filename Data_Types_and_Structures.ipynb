Q2.   Explain the difference between mutable and immutable data types with examples.
- Mutable data types are those whose values can be modified after creation. This means you can change, add, or remove elements without creating a new object. Examples of mutable data types include lists, dictionaries, and sets in Python.
Example: A list like [1, 2, 3] can be changed by adding another element, such as [1, 2, 3, 4].

- Immutable data types are those whose values cannot be changed once they are created. If you try to modify them, a new object is created instead of changing the original one. Examples include integers, floats, strings, and tuples.
Example: A string like "Hello" cannot be changed directly; if you modify it, a new string such as "Hello!" is created.

Q3. What are the main differences between lists and tuples in Python3?
- Lists and tuples are both used to store multiple items in a single variable in Python, but they have some important differences.

- Mutability  
Lists are mutable, which means their elements can be changed after creation.
Tuples are immutable, which means their elements cannot be changed once they are created.

- Syntax  
Lists are written using square brackets [ ].
Example: [1, 2, 3, 4]
Tuples are written using parentheses ( ).
Example: (1, 2, 3, 4)

- Performance   
Tuples are generally faster and use less memory because they are immutable, while lists are slightly slower due to their flexibility.

- Usage   
Lists are used when the data may need to change.
Tuples are used when the data should remain constant and should not be modified.

Q8.  What is a hash table, and how does it relate to dictionaries in Python?
- hash table is a data structure that stores data using key–value pairs. It uses a special function called a hash function to convert a key into a number (called a hash value), which helps quickly find the location where the value is stored.

- In Python, dictionaries are implemented using hash tables. This means that when you store data in a dictionary, Python uses a hash function to determine where the value should be placed in memory.

- For example:
{"name": "Rahul", "age": 17}
so, “name” and “age” are keys, and Python uses hashing to quickly locate their corresponding values.

Q13.  How do sets handle duplicate values in Python?
- sets do not allow duplicate values. When duplicate elements are added to a set, Python automatically removes them and keeps only one unique copy of each value.

This happens because sets are designed to store only unique elements. If the same value is added multiple times, it will appear only once in the set.

Q14.  How does the “in” keyword work differently for lists and dictionaries?
- the in keyword is used to check whether an element exists in a collection, but it works slightly differently for lists and dictionaries.

- For a list, the in keyword checks whether a value is present in the list.
Example:
5 in [1, 2, 3, 4, 5] → True because 5 is an element of the list.

- For a dictionary, the in keyword checks whether a key exists in the dictionary, not the value.
Example:
"name" in {"name": "Rahul", "age": 17} → True because "name" is a key.

Q16.  What is a nested dictionary, and give an example of its use case?
- a nested dictionary is a dictionary that contains another dictionary as its value. This means one dictionary is stored inside another dictionary.
Nested dictionaries are useful when we need to store complex or structured data.

```python
#example

students = {
  "student1": {"name": "Rahul", "age": 17},
  "student2": {"name": "Aman", "age": 18}
}
print(students)
```

- Here, the main dictionary stores information about different students, and each student has another dictionary containing their name and age.

- Use case: Nested dictionaries are commonly used to store data like student records, employee details, or product information, where each item has multiple related attributes.

Q17.  Describe the time complexity of accessing elements in a dictionary.
- This happens because dictionaries are implemented using a hash table. When we use a key to access a value, Python quickly converts the key into a hash value and directly finds the location of the data in memory.

- For example:
data = {"name": "Rahul", "age": 17}
Accessing data["age"] takes O(1) time in most cases.
However, in rare situations such as hash collisions, the time complexity can become O(n), but this does not usually happen.

Q19.Why are dictionaries considered unordered, and how does that affect data retrieval?
- dictionaries are traditionally considered unordered because their elements are not stored based on a fixed position like lists. Instead, the data is stored using hashing, which places items in memory based on their keys.
Because of this, the elements in a dictionary are not accessed by index. Instead, they are retrieved using their keys.

- For example:
data = {"name": "Rahul", "age": 17}

  To get the age, we use the key:
  data["age"]

  This method allows very fast data retrieval, because Python directly finds the value using the key rather than searching through all elements.



Q20. Explain the difference between a list and a dictionary in terms of data retrieval.
- the main difference between a list and a dictionary in terms of data retrieval is how the elements are accessed.

- In a list, elements are retrieved using their index (position). Each item has a specific position starting from 0.
Example:
numbers = [10, 20, 30]
To get the second element: numbers[1] → 20.

- In a dictionary, elements are retrieved using keys instead of positions. Each value is associated with a unique key.
Example:
data = {"name": "Rahul", "age": 17}
To get the age: data["age"] → 17.
