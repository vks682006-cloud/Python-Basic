Q22 What is the difference between @staticmethod and @classmethod in Python?
*   The **`@staticmethod`** and **`@classmethod`** decorators are used to define special methods in a class, but they differ in what they receive as their first argument. A **`@staticmethod`** does **not** take `self` or `cls` as a parameter and is used for utility functions that do not access or modify class or instance data. A **`@classmethod`**, on the other hand, takes `cls` (the class itself) as its first parameter and can access or modify class variables. In short, **`@staticmethod`** is used for independent helper methods, while **`@classmethod`** is used for methods that work with the class rather than individual objects.

Q1  Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog

Q2 Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Rectangle class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Creating objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Displaying areas
print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

Q3  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.

```python
# Parent class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

# Child class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

# Grandchild class
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery

    def display(self):
        print("Vehicle Type:", self.type)
        print("Brand:", self.brand)
        print("Battery:", self.battery)

# Creating an object
ecar = ElectricCar("Car", "Tesla", "75 kWh")

# Displaying details
ecar.display()
```

Q4  Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.




```python
# Base class
class Bird:
    def fly(self):
        print("Birds can fly.")

# Derived class
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky.")

# Derived class
class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly.")

# Creating objects
birds = [Sparrow(), Penguin()]

# Demonstrating polymorphism
for bird in birds:
    bird.fly()
```

Q5  Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.

```python
# BankAccount class
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private attribute

    # Deposit method
    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    # Withdraw method
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance!")

    # Check balance method
    def check_balance(self):
        print("Current Balance:", self.__balance)

# Creating an object
account = BankAccount(5000)

# Performing operations
account.check_balance()
account.deposit(2000)
account.check_balance()
account.withdraw(1500)
account.check_balance()
account.withdraw(6000)
```

Q6  Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar

```python
# Base class
class Instrument:
    def play(self):
        print("Instrument is playing.")

# Derived class
class Guitar(Instrument):
    def play(self):
        print("Guitar is playing: Strum Strum!")

# Derived class
class Piano(Instrument):
    def play(self):
        print("Piano is playing: Ting Ting!")

# Creating objects
guitar = Guitar()
piano = Piano()

# Runtime polymorphism
instruments = [guitar, piano]

for instrument in instruments:
    instrument.play()
```

method subtract_numbers() to subtract two numbers.

```python
# Class definition
class MathOperations:

    # Class method
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    # Static method
    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Calling the methods
print("Addition:", MathOperations.add_numbers(10, 5))
print("Subtraction:", MathOperations.subtract_numbers(10, 5))
```

Q10  Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.

```python
# Class definition
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Overriding __str__ for easy display
    def __str__(self):
        return f"({self.x}, {self.y})"

# Creating vector objects
v1 = Vector(2, 3)
v2 = Vector(4, 5)

# Adding vectors
v3 = v1 + v2

# Displaying results
print("Vector 1:", v1)
print("Vector 2:", v2)
print("Sum of Vectors:", v3)
```

Q14  Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
# Base class
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate salary
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

# Derived class
class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    # Overriding the salary calculation
    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

# Creating objects
employee = Employee(40, 500)
manager = Manager(40, 500, 5000)

# Displaying salaries
print("Employee Salary:", employee.calculate_salary())
print("Manager Salary:", manager.calculate_salary())
```

Q16 Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.

```python
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

# Derived class
class Cow(Animal):
    def sound(self):
        print("Cow says: Moo")

# Derived class
class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa")

# Creating objects
cow = Cow()
sheep = Sheep()

# Calling methods
cow.sound()
sheep.sound()
```

Q18  Create a class House with attributes address and price. Create a derived class Mansion that adds an

```python
# Parent class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

# Child class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def display(self):
        print("Address:", self.address)
        print("Price:", self.price)
        print("Number of Rooms:", self.number_of_rooms)

# Creating an object
mansion = Mansion("123 Green Street", 5000000, 12)

# Displaying details
mansion.display()
```
