Q1.  What is the difference between a function and a method in Python?


*    Function:
      A function is a standalone piece of code that performs a specific task. It is not inherently tied to any object or class. Functions take input (arguments), process them, and return output. They exist independently in the global scope or within modules.
- Method:
A method is essentially a function that is defined inside a class and is associated with objects created from that class. Methods often operate on the data (attributes) of the object they belong to. In Python, instance methods typically take self as their first parameter, which refers to the specific object calling the method.


```python


# Function
def square(number):
    return number * number

print(square(4))

#Method
class MathTool:
    def __init__(self, value):
        self.value = value

    def square(self):
        return self.value * self.value

tool = MathTool(4)
print(tool.square())
```

Q5.What are iterators in Python and how do they differ from iterables?
*    An iterable is an object that can be looped over, while an iterator is an object that produces one item at a time during iteration.

*  Iterable: Contains data and can be converted into an iterator using the iter() function. Examples: lists, tuples, strings, and dictionaries.
* Iterator: Keeps track of its current position and returns the next element using the next() function. Once all elements are returned, it raises StopIteration.


```python
#Example
# Iterable
numbers = [10, 20, 30]

# Create an iterator from the iterable
it = iter(numbers)

print(next(it))   # 10
print(next(it))   # 20
print(next(it))   # 30
```

Q6.Explain the concept of generators in Python and how they are defined.
*  Generators in Python:-

    A generator is a special type of function that produces values one at a time instead of returning all values at once. It uses the yield keyword instead of return, making it memory-efficient for handling large amounts of data.
* How Generators are Defined
   * A generator is defined like a normal function using def.
   * Instead of return, it uses yield to produce values one by one.
  * Each time next() is called, the generator resumes execution from where it last stopped





```python
#example
# Generator function
def numbers():
    yield 1
    yield 2
    yield 3

# Create generator object
gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

Q7.What are the advantages of using generators over regular functions?
*  Advantages of Generators over Regular Functions

   *  A generator is more efficient than a regular function because it produces values one at a time using the yield keyword instead of returning all values at once.
*  Advantages


1.   Memory Efficient: Generates values one by one instead of storing all values in memory.

1.   Faster Execution: Starts producing values immediately without waiting for the entire sequence
2.   Useful for Large Data: Can process large files or datasets without loading everything into memory


2.   Easy to Implement: Makes creating iterators simple using the yield keyword



```python
#Example
# Generator function
def numbers():
    for i in range(1, 6):
        yield i

# Create generator object
gen = numbers()

for num in gen:
    print(num)
```

* Syntax:

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
*  Difference between map(), filter(), and reduce() in Python

    map(), filter(), and reduce() are built-in functions used for processing collections of data.
  1. map()

     * Applies a function to every element in an iterable (list, tuple, etc.).
     * Modifies or transforms each element.
     * Returns a modified iterable (map object).
     * Example use: Doubling numbers or converting strings to uppercase.
2. filter()

     * Selects only the elements that satisfy a given condition.
     * Uses a function that returns True or False.
     * Removes elements that do not meet the condition.
     * Returns a filtered iterable (filter object).
     * Example use: Finding even numbers or positive values.
3. reduce()

     * Combines all elements of an iterable into a single value.
     * Applies a function repeatedly to accumulate the result.
     * Requires the functools module (from functools import reduce).
     * Returns a single result.
     * Example use: Calculating the sum or product of all numbers in a list.
   

```python
#Example

#map()
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, numbers))

print(result)


#filter()
numbers = [1, 2, 3, 4, 5, 6]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)


#reduce()
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
# Function to return the sum of all even numbers in a list
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example
my_list = [10, 15, 20, 25, 30, 35]
print("Sum of even numbers:", sum_even(my_list))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
# Function to check if a number is prime
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Print all prime numbers from 1 to 200
print("Prime numbers from 1 to 200 are:")

for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
# Iterator class for Fibonacci sequence

class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example
fib = Fibonacci(10)

print("Fibonacci sequence:")
for num in fib:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
# Generator function to yield powers of 2
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example
exp = 5

print("Powers of 2:")
for value in powers_of_two(exp):
    print(value, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
# Generator function to read a file line by line
def read_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

# Example
for line in read_file("sample.txt"):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of temperatures in Celsius
celsius = [0, 20, 30, 40, 100]

# Use map() to convert temperatures
fahrenheit = list(map(celsius_to_fahrenheit, celsius))

print("Temperatures in Fahrenheit:")
print(fahrenheit)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
# Function to check for non-vowel characters
def remove_vowels(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
text = "Hello World"

# Use filter() to remove vowels
result = "".join(filter(remove_vowels, text))

print("String after removing vowels:")
print(result)
```
