Q3.  What are pre-defined keywords in Python ?

  - Pre-defined keywords in Python are special words that the language has already reserved for specific tasks. You can’t use them as variable names because Python relies on them to understand what your code is supposed to do. Words like if, for, while, def, and class are all examples:- they’re basically the instructions that tell Python how to run your program.
Each keyword has a specific role. For example, "if" is used for decision-making, "for" and "while" are used for loops, and "def" is used to define functions.

Q5.  What is mutability in Python ?

  - Mutability is all about whether something can change after it’s created. In Python, if an object is mutable, you can update or modify it directly. If it’s immutable, you can’t change it — you’d have to make a new one.
For example:
- Lists are mutable — you can add, remove, or change items.
- Strings and tuples are immutable — once they’re made, they stay the same

______________________________________________________________________________

Q6.  Why are lists mutable, but tuples are immutable ?

- Lists are designed to be flexible — you can add, remove, or change items anytime. That’s super useful when your data needs to grow or shift. Python lets you do this by keeping the list’s structure open to changes.
- Tuples, on the other hand, are meant to be fixed. Once you create a tuple, it’s locked in. This makes them safer and faster in some cases, especially when you want to make sure the data doesn’t accidentally get changed.

-> So it’s really about purpose:

-	Lists are for dynamic, changeable data.
-	Tuples are for stable, unchanging data.

Q7. What is the difference between “==” and “is” operators in Python?

- The "==" operator checks if two values are equal — like, do they look the same.
- == → Are the contents the same?
- The is operator checks if two variables point to the exact same object in memory.
Here’s a simple way to think about it:
- is → Are they literally the same object?



```python
#Q7.
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True → same contents
print(a is b)  # False → different objects
```

Q8. What are logical operators in Python ?

- Logical operators are used to make decisions in your code — they help you combine conditions and control the flow based on whether something is true or false.
Python has three main logical operators:
- "and" → Returns True if both conditions are true
- 'or" → Returns True if at least one condition is true
- "not" → Flips the result — turns True into False, and vice versa


```python
#Q8.
x = 10
y = 5

print(x > 5 and y < 10)  # True — both are true
print(x < 5 or y < 10)   # True — one is true
print(not(x > y))        # False — x > y is True, so not makes it False
```

Q10.  What is the difference between implicit and explicit type casting?

- Type casting means converting one data type into another — like turning a float into an int, or a string into a number. There are two ways Python does this:
- Implicit Type Casting
Python does it automatically when it makes sense.
You don’t have to write anything — Python just handles it.
- Explicit Type Casting
You do it manually using functions like int(), float(), or str().
- In short:
- Implicit → Python decides for you
- Explicit → You tell Python what to do




```python
#Implicit type casting:-
x = 5       # int
y = 2.0     # float
z = x + y   # Python converts x to float automatically
print(z)    # Output: 7.0

#explicit type casting:-
age = "25"          # string
age_num = int(age)  # you convert it yourself
print(age_num + 5)  # Output: 30
```

Q11. What is the purpose of conditional statements in Python ?

- Conditional statements let your code make decisions. They help your program choose what to do based on certain conditions — like checking if a number is positive, or if a user is logged in.
The most common ones are if, elif, and else. You use them to run different blocks of code depending on what’s true at the moment.


```python
#Q11.
age = 18

if age >= 18:
    print("You're an adult.")
else:
    print("You're still a minor.")
```

Q12. How does the elif statement work ?

- The elif statement stands for “else if.” It’s used when you want to check multiple conditions, one after the other. If the first if isn’t true, Python moves on to check the elif. If that’s not true either, it keeps going — and finally, it’ll run the else block if nothing matched.
It helps keep your code clean instead of stacking a bunch of separate if statements.


```python
#Q12.
score = 75

if score >= 90:
    print("Excellent")
elif score >= 70:
    print("Good")
else:
    print("Keep trying")
```

Q13.  What is the difference between for and while loops ?

-  "for loop" is used when you know exactly how many times you want to repeat something.
- "while loop" is used when you want to keep going until a condition changes.



```python
#Q13.
#for loop
for i in range(5):
    print(i)
    # - This runs 5 times, from 0 to 4. Python handles the counting.
#while loop
i = 0
while i < 5:
    print(i)
    i += 1
    #- This also runs 5 times, but you control the condition and update manually.
```

Q14.  Describe a scenario where a while loop is more suitable than a for loop.

- Let’s say you’re building a program that asks the user to enter a password. You want to keep asking until they get it right.

```python
#Q14.
password = ""
while password != "python123":
    password = input("Enter your password: ")
```

Why use while here?
- You don’t know how many attempts the user will need.
- A "for" loop wouldn’t make sense because the number of tries isn’t fixed.
- The loop should continue until the condition (password == "python123") is met.

```python
#Q6
def change_list(my_list):
    my_list.append("new item")  # This will change the original list

def change_number(my_number):
    my_number += 10  # This won't affect the original number

# Try it out
original_list = [1, 2, 3]
original_number = 5

change_list(original_list)
change_number(original_number)

print("List after function:", original_list)       # List gets updated
print("Number after function:", original_number)   # Number stays the same
```

```python
#Q14.
# Ask the user to enter a string
text = input("Enter a string to reverse: ")

# Start from the last character
i = len(text) - 1
reversed_text = ""

# Keep adding characters from the end
while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
#Q15.
# Ask the user for a number
num = int(input("Enter a number to find its factorial: "))

# Start with 1 (since factorial of 0 is 1)
result = 1
n = num

# Multiply down to 1
while n > 0:
    result *= n
    n -= 1

print(f"Factorial of {num} is:", result)
```
